Load a dataset and plot the Almetrics Attention Score and the number of citations for the articles published in Journal of Sports Sciences between 2015 and 2019. 

```python
df = pd.read_csv ("JSSAltmetricCitations.csv", sep= ",")
print(df.head())
```

```python
dx="Altmetric Attention Score"; 
dy="Number of Dimensions citations"; 

ax = df.plot.scatter(x=dx, y=dy,c='black')

# Customize the plot
ax.set_xlabel('Altmetric Attention score', fontsize=16)
ax.set_ylabel('Nr. of citations', fontsize=16)     
ax.set_title('J of Sports Sci publications between 2015-2019', fontsize=16) 
ax.tick_params(labelsize=14)              
ax.set_xlim(0, None)
ax.set_ylim(0, None)
ax.grid(True, linestyle='--',c='grey', alpha=0.15)
ax.tick_params(axis='both') 

if savefigs:
    plt.savefig('figs/JSSAltmetricCitationsFig.pdf', bbox_inches='tight')
    plt.savefig('figs/JSSAltmetricCitationsFig.png', dpi=1200)
```
